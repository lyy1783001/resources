```python
# 辅助绘图函数

import matplotlib.pyplot as plt
import numpy as np

def plot_function(func, x_range, title="Function Plot", xlabel="x", ylabel="y", num_points=1000):
    # 在指定范围内生成num_points个点
    x = np.linspace(x_range[0], x_range[1], num_points)
    
    # 计算对应的y值
    y = func(x)
    
    # 创建图像并绘制曲线
    plt.figure(figsize=(8, 6))  # 设置图像大小
    
    # 直接使用Pynp张量绘图
    plt.plot(x, y, label=f"{getattr(func, '__name__', 'f(x)')}")
    
    # 添加标题和坐标轴标签
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    # 显示网格
    plt.grid(True, alpha=0.3)
    
    # 添加图例
    plt.legend()
    
    # 自动调整布局
    plt.tight_layout()
    
    # 显示图像
    plt.show()
```

## sigmoid

### 公式
$$
sigmoid(x) = \frac{1}{1+e^{-x}}
$$

```python
def sigmoid(x):
    """定义Sigmoid函数"""
    return 1 / (1 + np.exp(-x))
# 定义sigmoid函数和x轴范围
func = sigmoid
x_range = (-10, 10)
# 调用函数绘制sigmoid函数
plot_function(func, x_range)
```

## tanh

### 公式：
$$tanh(x) =  \frac{e^x - e^{-x}}{e^x + e^{-x}}$$

```python
def tanh(x):
    """定义函数"""
    exp_x = np.exp(x)
    exp_neg_x = np.exp(-x)
    return (exp_x - exp_neg_x) / (exp_x + exp_neg_x)
# 定义sigmoid函数和x轴范围
func = tanh
x_range = (-10, 10)
# 调用函数绘制sigmoid函数
plot_function(func, x_range)
```

## ReLU
### 公式
$$ReLU(x)=\max(0, x)$$

```python
def ReLU(x):
    """定义函数"""
    return np.maximum(0, x)
# 定义sigmoid函数和x轴范围
func = ReLU
x_range = (-10, 10)
# 调用函数绘制sigmoid函数
plot_function(func, x_range)
```
